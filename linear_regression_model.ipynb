# Linear regression model

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
real_x = np.linspace(-10,10,num=100)
real_y = 3 * real_x + 20
real_y += np.random.randn(len(real_x))*5 # noise
```

```python
plt.scatter(real_x,real_y)
plt.show()
```

```python
# create a model y = w*x to fit
tf.reset_default_graph()

x = tf.placeholder(dtype=tf.float32,shape=(None))
y = tf.placeholder(dtype=tf.float32,shape=(None))

w = tf.Variable(1, dtype=tf.float32)
b = tf.Variable(1, dtype=tf.float32)
pred = w*x + b

loss = tf.reduce_mean((y-pred)**2)

opt = tf.train.GradientDescentOptimizer(0.001)
update = opt.minimize(loss)
```

```python
# initial session
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# train model
for i in range(2000):
    l, _ = sess.run([loss, update], feed_dict={x:real_x, y:real_y})
    print(sess.run([w,b]))
```

```python
pred_y = sess.run(pred, feed_dict={x:real_x})

plt.scatter(real_x, real_y)
plt.plot(real_x, pred_y,'r')

plt.show()
```
